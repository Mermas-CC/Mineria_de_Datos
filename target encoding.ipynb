```python
categorical_data=data.drop(['FECHA_CORTE'],axis=1)
```

```python
categorical_data["POTENCIA_CONTRATADA"]=categorical_data["POTENCIA_CONTRATADA"].astype("category")
categorical_data["POTENCIA_CONTRATADA"]=categorical_data["POTENCIA_CONTRATADA"].cat.codes

mont_means=categorical_data.groupby("TARIFA")["POTENCIA_CONTRATADA"].mean()
mont_means
```

```python
categorical_data["TARIFA"]=categorical_data["TARIFA"].astype("category")
categorical_data["TARIFA"]=categorical_data["TARIFA"].cat.codes

mont_means=categorical_data.groupby("POTENCIA_CONTRATADA")["TARIFA"].mean()
mont_means
```
