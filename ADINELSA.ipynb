```python
data["UUID"] = data["UUID"].astype('category')
data["DEPARTAMENTO"] = data["DEPARTAMENTO"].astype('category')
data["PROVINCIA"] = data["PROVINCIA"].astype('category')
data["DISTRITO"] = data["DISTRITO"].astype('category')
data["LOCALIDAD"] = data["LOCALIDAD"].astype('category')
data["TARIFA"] = data["TARIFA"].astype('category')
data["FASE"] = data["FASE"].astype('category')
data["CODIGO_CIU"] = data["CODIGO_CIU"].astype('category')
print (data.dtypes)
```

```python
categorical_data=data.drop(['FECHA_CORTE','PERIODO_FACTURACION','UBIGEO','FECHA_INICIO','CONSUMO_KW','POTENCIA_CONTRATADA','MONTO_SOLES'],axis=1)
categorical_data.head()
```

```python
numerical_data=data.drop(['UUID','DEPARTAMENTO','PROVINCIA','DISTRITO','LOCALIDAD','TARIFA','FASE','CODIGO_CIU'],axis=1)
numerical_data.head()
```

```python
new_data=categorical_data.join(numerical_data)
new_data.head()
```

```python
encoded_dep= pd.get_dummies(categorical_data['DEPARTAMENTO'], prefix='DEPARTAMENTO')
encoded_prov= pd.get_dummies(categorical_data['PROVINCIA'], prefix='PROVINCIA')
encoded_dist= pd.get_dummies(categorical_data['DISTRITO'], prefix='DISTRITO')
encoded_local= pd.get_dummies(categorical_data['LOCALIDAD'], prefix='LOCALIDAD')
encoded_tarif= pd.get_dummies(categorical_data['TARIFA'], prefix='TARIFA')
encoded_fase= pd.get_dummies(categorical_data['FASE'], prefix='FASE')
encoded_codigo_ciu= pd.get_dummies(categorical_data['CODIGO_CIU'], prefix='CODIGO_CIU')
hot_cat_data1=categorical_data.join(encoded_dist)
hot_cat_data2=hot_cat_data1.join(encoded_prov)
hot_cat_data3=hot_cat_data2.join(encoded_dep)
hot_cat_data4=hot_cat_data3.join(encoded_local)
hot_cat_data5=hot_cat_data4.join(encoded_tarif)
hot_cat_data6=hot_cat_data5.join(encoded_fase)
hot_cat_data_fin=hot_cat_data6.join(encoded_codigo_ciu)

columna_dep = new_data['DEPARTAMENTO']
valores_unicos = columna_dep.unique()
print(valores_unicos)
```
